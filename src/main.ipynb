## Pyspark vs SQL

```python
%sql
SELECT * FROM samples.bakehouse.sales_customers
LIMIT 5
```

```python
sales_customer = spark.read.table("samples.bakehouse.sales_customers").limit(5)
df_selected = sales_customer.select("customerID", "first_name", "last_name", "email_address", "phone_number", "country", "gender")
display(df_selected)
```

```python
%sql
SELECT DISTINCT country
FROM samples.bakehouse.sales_customers
```

```python
df_distinct = df_selected.select("country").distinct()
display(df_distinct)
```

```python
%sql

SELECT *
FROM samples.bakehouse.sales_customers
WHERE 1=1
AND country = 'Japan'
AND gender = 'female'
LIMIT 5
```

```python
from pyspark.sql.functions import col
df_where = df_selected.where(
    (col("country") == "Japan") & 
    (col("gender") == "female")
    )
display(df_where)
```

```python
from pyspark.sql.functions import concat, lit

df_plus_columns= df_selected.select(
    concat(col("first_name"), lit(" "), col("last_name")).alias("full_name"),
    lit(True).alias("is_tue"),
    lit("random comment").alias("comment")
)

display(df_plus_columns)
```
